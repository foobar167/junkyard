**Contrastive learning** is a machine learning technique used to learn meaningful representations from data by contrasting similar and dissimilar examples. It aims to train a model to differentiate between instances, pulling representations of similar pairs closer together while pushing dissimilar pairs further apart in a feature space. This approach is particularly useful for **self-supervised learning**, where labeled data is scarce or unavailable.
